## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

#### 1.- Write a function that generates uniformly distributed random numbers.

```python
# your code here
def ran_uni_dist(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    a = bottom
    b = ceiling
    randoms  = a + (b-a)*x
    return randoms
```

#### 2.- Call your function with 2 sets of params defined

```python
bottom=10
ceiling=15
count=100

set1 = ran_uni_dist(bottom, ceiling, count)
print(set1)
```

```python
bottom=10
ceiling=60
count=1000

set2 = ran_uni_dist(bottom, ceiling, count)
print(set2)
```

#### 3.- Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

```python
bins=10
plt.hist(set1, bins=bins) 
plt.show()
```

```python
bins=10
plt.hist(set2, bins=bins)  
plt.show()
```

How are the two distributions different?

**The higher is the count, the flattern is the histogram**

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

#### 1.-Write a function that generates normally distributed random numbers.

```python
from scipy.stats import norm
def ran_norm_dist(average, deviation, count):
    randoms = np.random.normal(average, deviation, count)
    return randoms
```

#### 2.-Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1

```python
average = 10
deviation = 1
count = 1000

set3 = ran_norm_dist(average, deviation, count)
print(set3)
```

#### 3.- Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50

```python
average = 10
deviation = 50
count = 1000

set4 = ran_norm_dist(average, deviation, count)
print(set4)
```

#### 4.- Plot the distributions of the data generated.

```python
bins=50
plt.hist(set3, bins=bins)  
plt.show()
```

```python
bins=50
plt.hist(set4, bins=bins)  
plt.show()
```

How are the two distributions different?

**In the first, the results tend to be more close to the mean**

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 1 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

```python
# your code here

def ran_norm_exp(mean, count):
    randoms = np.random.exponential(1/mean, count)
    return randoms
 
```

```python
mean = 10
count = 1

set5 = ran_norm_exp(mean, count)
print(set5)
```

```python
mean = 10
count = 100

set6 = ran_norm_exp(mean, count)
print(set6)
```

```python
bins=100
plt.hist(set5, bins=bins)  
plt.show()
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# Hint: This is same as saying P(x<15)
lamb = 10
x = 15

p = 1 - np.exp(-lamb*x)
print(p)
```

```python
from scipy.stats import expon
expon.cdf(x, loc=0, scale=1/lamb)
```

What is the probability that the customer will spend more than 15 minutes

```python
print('The probability the customer will spend more than 15 minutes is', round(1-p,3))
```
